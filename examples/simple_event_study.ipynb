# Event Study

```python
import pandas as pd
from event_studies import EventStudy, EventStudyConfig
```

```python
events = pd.read_csv('data_input/events.csv')
events['event_date'] = pd.to_datetime(events['event_date'])

returns = pd.read_csv('data_input/returns.csv', index_col='date', parse_dates=True)
for col in returns.columns:
    returns[col] = pd.to_numeric(returns[col], errors='coerce')

factors = pd.read_csv('data_input/factors.csv', index_col='date', parse_dates=True)
```

```python
config = EventStudyConfig(model='market_model',
                          estimation_window=(-60, -11), 
                          event_window=(-5, 5), 
                          alpha=0.05)

study = EventStudy(config)

results = study.run(events=events, returns=returns, factors=factors)
```
